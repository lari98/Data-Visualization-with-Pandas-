```python
# Import the necessary libraries
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:/Anaconda/data.csv")
```

```python
# visualize the first 3 rows of the dataset
df.head(3) # df.head(number_of_rows)
```

```python
# For the sake of this tutorial we are going to analyse the first 11 rows in the dataset 
df = df.head(11)
```

```python
# Line Plot - to do a line plot we will use the pandas's dataframe's function plot()
df.plot() # or df.plot.line()

plt.title('Line Plot') # add title to the plot
plt.xlabel('X') # add a label for the x axis
plt.ylabel('Y') # add a label for the y axis
plt.show()
```

```python
# Scatter plot - to do this plot we will use the sintax: plot(kind='scatter',x,y) or plot.scatter(x,y)
df.plot(kind='scatter', x='Duration', y='Calories') # df.plot.scatter(x='Duration', y='Calories')

plt.title('Scatter Plot') 
plt.xlabel('Duration') 
plt.ylabel('Calories')
plt.show()
```

```python
# Histogram - the syntax for the histogram is: plot(kind='hist') or plot.hist()
df.plot(kind='hist') # df.plot.hist()

plt.title('Histogram') 
plt.show()
```

```python
# Area Plot - syntax: plot(kind='area') orr plot.area()
df.plot(kind='area') # df.plot.area()

plt.title('Area Plot') 
plt.show()
```

```python
# Bar Chart - syntax: plot(kind='bar') or plot.bar()
df.plot(kind='bar') # df.plot.bar()

plt.title('Bar Chart') 
plt.show()
```

```python
# Stacked Bar Plot - syntax: plot(kind='bar', stacked=True) or bar(stacked=True)
df.plot(kind='bar', stacked=True) # df.plot.bar(stacked=True)

plt.title('Stacked Bar Chart') 
plt.show()
```

```python
# Horizontal bar - plot(kind='barh') or barh()
df.plot(kind='barh') # df.plot.barh()

plt.title('Horizontal Bar Chart') 
plt.show()
```

```python
# Box Plot - syntax: plot(kind='box') ot box()
df.plot(kind='box') # df.plot.box()

plt.title('Box Plot') 
plt.show()
```

```python
# Hexagonal binning plot - syntax: plot(kind='hexbin',x,y) or plot.hexbin(x,y)
df.plot(kind='hexbin', x='Maxpulse', y='Calories', gridsize=18) # df.plot.hexbin(x='Maxpulse', y='Calories', gridsize=18)

plt.title('Hexagonal Binning Plot') 
plt.show()
```

```python
# Density Estimate plot or Kernel density estimate charts - syntax: plot(kind='kde') / plot(kind='density') or plot.kde() / plot.density()
df.plot(kind='kde') # df.plot(kind='density') or df.plot.kde() or df.plot.density()

plt.title('Density Estimate Plot') 
plt.show()
```

```python
# Pie Plot - syntax: plot(kind='pie',y) or plot.pie(y)
df.plot(kind='pie',y='Duration',autopct='%1.1f%%')

plt.title('Pie Plot') 
plt.show()
```

```python
# Save the plot as an image - savefig()
df.plot()
plt.title('Line Plot') 
plt.show()

plt.savefig('lineplot.png') 
```
